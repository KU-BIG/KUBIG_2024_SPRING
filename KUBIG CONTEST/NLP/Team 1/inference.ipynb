```python
import pandas as pd
import re

# 데이터 불러오기
df = pd.read_csv('./data/test.csv')

# 새로운 열을 담을 리스트 초기화
question_1 = []
question_2 = []

# 각 행을 순회하며 질문 분리
for index, row in df.iterrows():
    # 정규표현식을 사용하여 '. 또한', '? 또한', '. 그리고', '? 그리고'를 기준으로 분리
    questions = re.split('\. 또한|\? 또한|\. 그리고|\? 그리고', row['질문'])
    
    # 첫 번째 질문은 question_1에 추가
    question_1.append(questions[0].strip())
    
    # 두 번째 질문이 있는 경우에는 question_2에 추가, 없는 경우에는 빈 문자열 추가
    if len(questions) > 1:
        # 두 번째 질문의 시작이 쉼표로 시작하는 경우 쉼표 제거
        question_2_text = questions[1].strip()
        if question_2_text.startswith(','):
            question_2_text = question_2_text[1:].strip()
        question_2.append(question_2_text)
    else:
        question_2.append('')

# 새로운 열 추가
df['질문_1'] = question_1
df['질문_2'] = question_2

# 결과 확인
print(df)
```

```python
from transformers import GPT2LMHeadModel, PreTrainedTokenizerFast, AdamW
from tqdm import tqdm
device = 'cpu'
```

```python
# 저장된 Fine-tuned 모델과 토크나이저 불러오기
model_dir = "kogpt2_학습데이터전처리/ckpt_ep_30"
model = GPT2LMHeadModel.from_pretrained(model_dir)
model.to(device)
tokenizer = PreTrainedTokenizerFast.from_pretrained(model_dir)

# Inference를 위한 test.csv 파일 로드
test = pd.read_csv('test2.csv')

# '질문_1'과 '질문_2' 컬럼을 함께 사용하여 답변 생성 및 저장
preds = []

for index, row in tqdm(test.iterrows(), total=len(test)):
    # 첫 번째 질문에 대한 답변 생성
    input_ids_1 = tokenizer.encode(row['질문_1'] + tokenizer.eos_token, return_tensors='pt')
    output_sequences_1 = model.generate(
        input_ids=input_ids_1.to(device),
        max_length=300,
        temperature=0.9,
        top_k=1,
        top_p=0.9,
        repetition_penalty=1.2,
        do_sample=True,
        num_return_sequences=1
    )
    for generated_sequence in output_sequences_1:
        full_text_1 = tokenizer.decode(generated_sequence, skip_special_tokens=False)
        answer_start_1 = full_text_1.find(tokenizer.eos_token) + len(tokenizer.eos_token)
        answer_only_1 = full_text_1[answer_start_1:].strip()
        answer_only_1 = answer_only_1.replace('\n', ' ')
    
    # 두 번째 질문에 대한 답변 생성 (질문2가 비어있는 경우 공백 문자열 생성)
    if pd.notnull(row['질문_2']): 
        input_ids_2 = tokenizer.encode(row['질문_2'] + tokenizer.eos_token, return_tensors='pt')
        output_sequences_2 = model.generate(
            input_ids=input_ids_2.to(device),
            max_length=300,
            temperature=0.9,
            top_k=1,
            top_p=0.9,
            repetition_penalty=1.2,
            do_sample=True,
            num_return_sequences=1
        )
        for generated_sequence in output_sequences_2:
            full_text_2 = tokenizer.decode(generated_sequence, skip_special_tokens=False)
            answer_start_2 = full_text_2.find(tokenizer.eos_token) + len(tokenizer.eos_token)
            answer_only_2 = full_text_2[answer_start_2:].strip()
            answer_only_2 = answer_only_2.replace('\n', ' ')
    else:
        answer_only_2 = ""  # 두 번째 질문이 비어있는 경우 빈 문자열 사용
    
    preds.append({'id': row['id'], 'answer_1': answer_only_1, 'answer_2': answer_only_2})

# 같은 id를 가진 행들의 답변 합치기
merged_preds = []
for id_value in test['id']:
    answers = [pred['answer_1'] + ' ' + pred['answer_2'] for pred in preds if pred['id'] == id_value]
    merged_preds.append({'id': id_value, 'merged_answer': ' '.join(answers)})

# 결과 확인
for merged_pred in merged_preds:
    print(merged_pred)
```

```python
preds = [text['merged_answer'].replace('<s>', '') for text in merged_preds]
```

```python
for item in merged_preds:
    item['merged_answer'] = item['merged_answer'].replace('?', '')  # 물음표 제거
```

```python
# Test 데이터셋의 모든 질의에 대한 답변으로부터 512 차원의 Embedding Vector 추출
# 평가를 위한 Embedding Vector 추출에 활용하는 모델은 'distiluse-base-multilingual-cased-v1' 이므로 반드시 확인해주세요.
from sentence_transformers import SentenceTransformer # SentenceTransformer Version 2.2.2

# Embedding Vector 추출에 활용할 모델(distiluse-base-multilingual-cased-v1) 불러오기
model = SentenceTransformer('distiluse-base-multilingual-cased-v1')

# 생성한 모든 응답(답변)으로부터 Embedding Vector 추출
pred_embeddings = model.encode(preds)
pred_embeddings.shape
```
