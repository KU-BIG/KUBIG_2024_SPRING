```python
def preprocess(data):
    csv = []
    for i in tqdm(range(len(data['documents']))):
        news = data['documents'][i]
        id = news['id']
        category = news['category']
        media_type = news['media_type']
        media_sub_type = news['media_sub_type']
        media_name = news['media_name']
        size = news['size']
        char_count = news['char_count']
        date = news['publish_date']
        title = news['title']
        texts = ''
        for idx, line in enumerate(news['text']):
            try:
                text = line[0]['sentence']
                texts += text
                texts += ' '
            except:
                None
        ext = news['extractive']
        summary = ''
        for j in ext:
            try:
                line = news['text'][j][0]['sentence']
                summary += line
                summary += ' '
            except:
                None
        abs = news['abstractive'][0]
        csv.append([id, category, media_type, media_sub_type, media_name, size, char_count, date, title, texts, summary, abs])
    return csv
```

```python
import pandas as pd
from tqdm import tqdm

model = BartForConditionalGeneration.from_pretrained(model_binary_path).cuda()

df = pd.read_csv('/home/yeobinhong/Downloads/news.csv')
df['summary'] = None

for i, text in (enumerate(tqdm(df['text'][:2000]))):
    input_text = ''
    for line in df['text'][i]:
        line = line.replace('\\','').replace('[','').replace(']','').replace('\'','').replace('\"','').replace(',','') # 정규표현식 써야하는데 죄송합니다 ...
        input_text += line
    if len(input_text) <= 10:
        continue
    input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(input_ids.cuda(), max_length=150, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary_text = tokenizer.decode(summary_ids[0].cpu(), skip_special_tokens=True)
    df.at[i, 'summary'] = summary_text
```
