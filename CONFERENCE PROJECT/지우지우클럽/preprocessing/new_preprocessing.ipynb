## Groupby 진행
유저 정보들로 groupby했을 때 크롤링한 시간 차이로 인해 다른 사람으로 인식되는 경우를 전처리
- 어떤 소비자가 어떤 버거집을 갔는지에 대한 정보 필요

```python
groupby = pd.DataFrame(df.groupby(['username', 'userid_r', 'userid_p','userid_f'])['restaurant'].agg(['unique','nunique','count'])).reset_index()
groupby
```

```python
# ***이 없는 중복된 이름들 추출
name_list = []
for i in range(len(groupby)):
    try:
        if groupby['username'][i] == groupby['username'][i+1]:
            if '****' not in groupby['username'][i]:
                name_list.append(groupby['username'][i])
    except:
        break

set(name_list)
```

```python
len(set(name_list))
```

```python
copy = df.copy()
max1 = df[df['username']=='옴뇸뇸ㅎ']['userid_r'].max()
copy.loc[copy['username']=='옴뇸뇸ㅎ', 'userid_r'] = max1
max2 = df[df['username']=='옴뇸뇸ㅎ']['userid_p'].max()
copy.loc[copy['username']=='옴뇸뇸ㅎ', 'userid_p'] = max2
copy[copy['username']=='옴뇸뇸ㅎ']
```

```python
# string 값을 float 값으로 바꾸기
df['userid_f'] = df['userid_f'].fillna(0)
df['userid_f'] = df['userid_f'].str.replace(',','')
df['userid_f'] = df['userid_f'].astype(float)
```

```python
# 중복값의 이름들에 대하여 최댓값으로 유저 고유정보 바꾸기
for name in set(name_list):
    max1 = df[df['username']==name]['userid_r'].max()
    max2 = df[df['username']==name]['userid_p'].max()
    max3 = df[df['username']==name]['userid_f'].max()
    df.loc[df['username']==name, 'userid_r'] = max1
    df.loc[df['username']==name, 'userid_p'] = max2
    df.loc[df['username']==name, 'userid_f'] = max3
```

# 리뷰 텍스트 전처리

```python
df = pd.read_csv('1차 리뷰 전처리(원본).csv', index_col=0)
df.head()
```

```python
df['text']
```

```python
df['text'].isnull().sum()
```

```python
df['text'].dropna()
```

```python
text_list = df['text'].dropna().tolist()
text_list
```

```python
# 이모티콘 제거
import emoji

for i in range(len(text_list)):
    try:
        text_list[i] = emoji.replace_emoji(text_list[i], replace='')
    except:
        continue
```

```python
# \n 엔터키 제거
text_list = list(map(lambda x: x.replace('\n', ' '), text_list))
```

```python
# ~ 제거
text_list = list(map(lambda x: x.replace('~', ''), text_list))
```

```python
text_list
```

1. 리뷰에서 영어표현, puncation, 이모티콘 모두 제거(=한글만 남기자.)
    - 이모티콘 지우기로 인해 공백이 된 문자들 => 전부 좋아요 표시이기 때문에 '좋아요'라고 바꾸기. 
2. 띄어쓰기 변환기를 통해 띄어쓰기가 되어있지 않은 문장들을 제대로 변환한다.
3. 형태소 분석기를 통해 각각의 리뷰를 형태소 단위로 분리한다. 
4. 한글자로 되어 있는 것 제거?
5. 불용어 제거
6. 최소 빈도수 이하의 것들은 제거
7. 잘못 분리된 의미있는 단어들을 찾아서 사용자 단어 사전에 추가. 

```python
for i in range(len(text_list)):
    if text_list[i] == '':
        text_list[i] = '좋아요.'

text_list
```

```python
file_name = 'text_list.txt'
with open(file_name, 'w') as file:
    file.write('\n'.join(text_list))
```
