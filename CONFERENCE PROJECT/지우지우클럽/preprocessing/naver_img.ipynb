```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

from urllib.request import urlretrieve
```

```python
option = Options()
option.add_argument('--headless')
driver = webdriver.Chrome(options=option)

def img_crawling(id):
    url = 'https://m.place.naver.com/restaurant/'+str(id)+'/menu/list'
    driver.get(url)
    time.sleep(3)

    li = driver.find_elements(By.CLASS_NAME, 'E2jtL')
    restaurant = driver.find_element(By.CLASS_NAME, 'Fc1rA').text

    # img 폴더 만든 후 경로 복사하기
    img_folder_path = '/Users/choejuhui/Desktop/img'

    for i in li:
        cache = i.text.split('\n')
        if '대표' in cache:
            continue
            
        try:
            img = i.find_element(By.CLASS_NAME, 'K0PDV')
            style = img.get_attribute('style')
            url = style.split('url')[-1][2:-3]
            urlretrieve(url, img_folder_path+'/{}_{}.jpg'.format(restaurant, cache[0]))
        
        except:
            continue

for num in id:
    img_crawling(num)
```
