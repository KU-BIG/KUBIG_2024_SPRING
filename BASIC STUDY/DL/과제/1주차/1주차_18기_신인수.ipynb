- 머신러닝 기초와 비슷한 내용이지만 파이토치로 실행해보며 익숙해지는것이 목표

```python
%matplotlib inline
import random
import torch
from d2l import torch as d2l
```

```python
def synthetic_data(w, b, num_examples):
    """Generate y = Xw + b + noise."""
    X = torch.normal(0, 1, (num_examples, len(w)))
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape((-1, 1))

true_w = torch.tensor([3, -1.5])
true_b = 5.5
features, labels = synthetic_data(true_w, true_b, 2000)
```

```python
X = list(features)
random.shuffle(X)
X
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    # The examples are read at random, in no particular order
    random.shuffle(indices)
    ###반복문으로 채워넣기:

    for i in range(0, num_examples, batch_size):

        ##이하는 반복문에 대한 조건입니다.
        batch_indices = torch.tensor(
            indices[i: min(i + batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices]
```

```python
batch_size = 20

for X, y in data_iter(batch_size, features, labels):
    print(X, '\n', y)
    break
```

```python
w = torch.normal(0, 0.01, size=(2,1), requires_grad=True)
b = torch.tensor([3.0], requires_grad = True)#원하는 실수를 넣어주세요
```

```python
def linreg(X, w, b):
    """The linear regression model."""
    return torch.matmul(X, w) + b
```

```python
def squared_loss(y_hat, y):
    """Squared loss."""
    loss = (y_hat - y)**2 / 2
    return loss.mean()
```

```python
# got help from chatgpt. unlikely to work
def sgd(params, lr, batch_size):
    """Minibatch stochastic gradient descent."""
    with torch.no_grad():
    ###반복문으로 채워넣기
        # Initialize gradients to zero
        for param in params:
            if param.grad is not None:
              param -= lr * param.grad / batch_size
              param.grad.zero_()

```

```python
lr = 0.01
num_epochs = 100
net = linreg
loss = squared_loss

for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y)
        # Compute gradient on `l` with respect to [`w`, `b`]
        l.sum().backward()
        sgd([w, b], lr, batch_size)
    with torch.no_grad():
        train_l = loss(net(features, w, b), labels)
        print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')
```

```python
print(f'error in estimating w: {true_w - w.reshape(true_w.shape)}')
print(f'error in estimating b: {true_b - b}')
```

### Concise Implementation of Linear Regression
- API 사용해보기

```python
import numpy as np
import torch
from torch.utils import data
from d2l import torch as d2l

true_w = torch.tensor([3, -1.5])
true_b = 5.5
features, labels = d2l.synthetic_data(true_w, true_b, 2000)
```

```python
def load_array(data_arrays, batch_size, is_train=True):
    """Construct a PyTorch data iterator."""
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset, batch_size, shuffle=is_train)

batch_size = 20
data_iter = load_array((features, labels), batch_size)
```

```python
next(iter(data_iter))
```

```python
from torch import nn

net = nn.Sequential(nn.Linear(2, 1))
```

```python
net[0].weight.data.normal_(0, 0.01)
net[0].bias.data.fill_(0)
```

```python
loss = nn.MSELoss() ###api function으로 직접 설정
```

```python
trainer = torch.optim.SGD(net.parameters(), lr = 0.01) ###api function으로 직접 설정
```

```python
for X, y in data_iter:
    print(X, y)
    break
```

```python
num_epochs = 3
for epoch in range(num_epochs):
    for X, y in data_iter:
        l = loss(net(X), y)
        trainer.zero_grad()
        l.backward()
        trainer.step()
    l = loss(net(features), labels)
    print(f'epoch {epoch + 1}, loss {l:f}')
```

```python
w = net[0].weight.data ###학습결과 w 구하는 식
print('error in estimating w:', true_w - w.reshape(true_w.shape))
b = net[0].bias.data ###학습결과 b 구하는 식
print('error in estimating b:', true_b - b)
```
