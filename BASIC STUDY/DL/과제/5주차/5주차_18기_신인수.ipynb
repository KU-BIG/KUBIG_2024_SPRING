```python
#1. 생성할 문장 데이터

sentence = ("Brick walls are there for a reason and you must not think "
            "that the brick walls aren't there to keep us out, but rather "
            "in this way that the brick walls are there to show us how badly we want things.")
```

```python
#2. 문자 집합 만들기
world_set = list(set(sentence))

world_set
```

```python
for x, y in enumerate(world_set):
  print(x, y)
```

```python
## 문제(1): 각 문자에 정수 인코딩 (공백도 하나의 원소로 포함)
# 원래 output과는 다르지만 이렇게 해보자
vocab = {char: idx for idx, char in enumerate(world_set)}
vocab
```

```python
print(vocab)
```

```python
#3. 문자 집합 크기 확인

vocab_size = len(vocab)
print('문자 집합 크기 : {}'.format(vocab_size))
```

```python
#4. 하이퍼 파라미터 설정(자유롭게 수정해보세요!)
# 일단 그대로 해보자

hidden_size = vocab_size # 같아야 하는 것 확인!
sequence_length = 10  # 너무 길거나 너무 짧게 잡으면 안됩니다!
learning_rate = 0.01
```

```python
#5. seqence 길이 단위 자르기

# 데이터 구성을 위한 리스트
x_data = []
y_data = []
```

```python
vocab["B"]
```

```python
# 내 풀이
for i in range(0, len(sentence) - sequence_length):
  x_str = sentence[i:i+sequence_length]
  y_str = sentence[i+1 : i+sequence_length+1]
  print(i, x_str, "->", y_str)

  # x_str과 y_str이 문자집합에 해당하는 인덱스를 각각 x_data, y_data에 append
  x_data.append([vocab[c] for c in x_str])
  y_data.append([vocab[d] for d in y_str])
```

```python
# 원래 결과
## 문제(2): 반복문 내에서의 인덱싱을 사용하여 sequence_length 값 단위로 샘플을 잘라 데이터 만들기, y_str은 x_str은 한 칸씩 쉬프트된 sequnce

for i in range(0, len(sentence) - sequence_length):
  x_str = sentence[##]
  y_str = sentence[##]
  print(i, x_str, "->", y_str)

  # x_str과 y_str이 문자집합에 해당하는 인덱스를 각각 x_data, y_data에 append
  x_data.append([vocab[c] for c in x_str])
  y_data.append([vocab[d] for d in y_str])
```

```python
# 출력해서 한 칸씩 쉬프트된 것 확인하기!
# shift 되었다.
print(x_data[0])
print(y_data[0])
```

```python
x_data
```

```python
##6. 입력 시퀀스에 대해 원핫인코딩 수행

## 문제(4) : x_data를 원핫인코딩 > numpy의 eye를 쓸 수 있지 않을까?
x_one_hot = [np.eye(vocab_size)[idx] for idx in x_data]
x_one_hot
```

```python
##7. 입력 데이터, 레이블데이터 텐서로 변환

## 문제(5) : x_one_hot과 y_data 텐서로 변환 : 둘 다 같은 형식의 텐서로 변환하면 될까?? (FloatTensor, LongTesor 중 맞는 것은?)
X = torch.FloatTensor(np.array(x_one_hot))
Y = torch.LongTensor(y_data)
```

```python
##8. 크기 확인
print('훈련 데이터의 크기 : {}'.format(X.shape))
print('레이블의 크기 : {}'.format(Y.shape))
```

```python
##9.원핫인코딩 결과 샘플 확인하기
print(X[0])
```

```python
##10. 레이블 데이터 샘플 확인하기
print(Y[0])
```

```python
##11. RNN 모델 구현

##문제(6) : 기본 pytorch 인자 넣기 연습 + forward 채우기
### 조건 : rnn layer 2개 쌓기 + 마지막은 fc layer
### batch_fisrt 설정 필요할까? (유튜브 강의 참고) 일단 True 로 하고 있긴 하다

class Net(torch.nn.Module):
  def __init__(self, input_dim, hidden_dim, layers):
    super(Net, self).__init__()

    # Define the first RNN layer
    self.rnn1 = nn.RNN(input_dim, hidden_dim, layers, batch_first=True)

    # Define the second RNN layer
    self.rnn2 = nn.RNN(hidden_dim, hidden_dim, layers, batch_first=True)

    # Define the fully connected layer
    self.fc = nn.Linear(hidden_dim, vocab_size)

  def forward(self, x):
    out1, _ = self.rnn1(x)

    # Forward pass through the second RNN layer
    out2, _ = self.rnn2(out1)

    # Fully connected layer
    out = self.fc(out2)

    return out
```

```python
net = Net(vocab_size, hidden_size, 2)
```

```python
##12. loss function
criterion = torch.nn.CrossEntropyLoss()

##13. optimizer
optimizer = optim.Adam(net.parameters(), learning_rate)
```

```python
##14. 출력 크기 점검
outputs = net(X)
print(outputs.shape)
```

```python
Y.shape
```

The CrossEntropyLoss in PyTorch expects the output tensor outputs to have the shape (batch_size, num_classes) and the target tensor Y to have the shape (batch_size,).

CrossEntropyLoss이 기대하는 입력값이 num_classes인데 이 경우 vocab_size에 해당

```python
##15. Training 시작

for i in range(100):
    optimizer.zero_grad()
    outputs = net(X)

    ##문제(7) : outputs, Y 형태 그대로 넣으면 안되죠. view 함수를 이용해 loss값을 계산해봅시다.
    loss = criterion(outputs.view(-1, vocab_size), Y.view(-1))
    loss.backward()
    optimizer.step()

    #16. 예측결과 확인
    results = outputs.argmax(dim=2)
    predict_str = ""
    for j, result in enumerate(results):
        if j == 0: # 처음에는 예측 결과를 전부 가져오기
            predict_str += ''.join([world_set[t] for t in result])
        else: # 그 다음에는 마지막 글자만 반복 추가
            predict_str += world_set[result[-1]]

    print(predict_str)
```
