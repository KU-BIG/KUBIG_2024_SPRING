# 3. 데이터 전처리
fasttext: subword 단위로 임베딩 벡터를 생성하는 도구. 한국어에서 subword는 자음과 모음이라고 할 수 있음.

fasttext에 학습시킬 데이터를 만들기 위해 앞서 로드한 네이버 쇼핑 리뷰들을 분해해 자음 모음 단위로 전처리 해보자.

```python
def word_to_jamo(token):
  def to_special_token(jamo): # 경우에 따라 초, 중, 종성이 다 있는 게 아닌 경우가 있음. => 이 때 -를 반환
    if not jamo:
      return '-'
    else:
      return jamo

  decomposed_token = ''
  for char in token:
    try:
      # 음절을 초성, 중성, 종성으로 분리
      cho, jung, jong = hgtk.letter.decompose(char)

      # 자모가 빈 문자일 경우 특수문자 -로 대체
      cho = to_special_token(cho)
      jung = to_special_token(jung)
      jong = to_special_token(jong)
      decomposed_token = decomposed_token + cho + jung + jong

    # 만약 음절이 한글이 아닐 경우 자모를 나누지 않고 추가
    except Exception as exception:
      if type(exception).__name__ == 'NotHangulException':
        decomposed_token += char

  # 단어 토큰의 자모 단위 분리 결과를 추가
  return decomposed_token
```

```python
print(word_to_jamo('남동생'))
print(word_to_jamo('야구')) # 야구는 종성이 없으므로 종성 부분을 -로 반환
```

```python
print(word_to_jamo('여동생'))
```

```python
print(mecab.morphs('선물용으로 빨리 받아서 전달했어야 하는 상품이었는데 머그컵만 와서 당황했습니다.'))
```

```python
# mecab으로 형태소 분리 후 그 형태소마다 각각 자음모음으로 분리해주는 함수
def tokenize_by_jamo(s):
  return [word_to_jamo(token) for token in mecab.morphs(s)]
```

```python
print(tokenize_by_jamo('선물용으로 빨리 받아서 전달했어야 하는 상품이었는데 머그컵만 와서 당황했습니다.'))
```

```python
# 리뷰 데이터의 reviews 컬럼만을 가져와서 자모 분리
tokenized_data = []

for sample in tqdm(total_data['reviews'].to_numpy()):
    tokenized_sample = tokenize_by_jamo(sample) # 자소 단위 토큰화
    tokenized_data.append(tokenized_sample)
```

자모 분리된 상태에서 역으로 다시 단어로 결합시키는 함수를 정의하자.

왜? => 코사인 유사도를 평가할 때 자모 분리된 상태가 아닌 단어 상태로 편하게 보기 위해서.

```python
def jamo_to_word(jamo_sequence):
  tokenized_jamo = []
  index = 0

  while index < len(jamo_sequence):
    if not hgtk.checker.is_hangul(jamo_sequence[index]):
      tokenized_jamo.append(jamo_sequence[index])
      index += 1

    else:
      tokenized_jamo.append(jamo_sequence[index:index+3])
      index += 3

  word = ''
  try:
    for jamo in tokenized_jamo:
      if len(jamo)==3:
        if jamo[2] == '-':
          word += hgtk.letter.compose(jamo[0], jamo[1])
        else:
          word += hgtk.letter.compose(jamo[0], jamo[1], jamo[2])
      else:
        word += jamo
  except Exception as exception:
    if type(exception).__name__ == 'NotHangulException':
      return jamo_sequence

  return word
```

```python
print(jamo_to_word('ㄴㅏㅁㄷㅗㅇㅅㅐㅇ'))
```

fasttext를 실행하기 전에 훈련 대상인 단어들을 txt 파일로 준비해둬야 함.

==> tokenized_data.txt라는 파일을 쓰기 모드(w)로 생성해주고 앞서 전처리한 tokenized_data를 입력해줌.

```python
# 가독성이 좋게 출력해보기
def transform(word_sequence):
  return [(jamo_to_word(word), similarity) for (similarity, word) in word_sequence]
```

```python
print(transform(model.get_nearest_neighbors(word_to_jamo('남동생'), k=10)))
print(transform(model.get_nearest_neighbors(word_to_jamo('택배'), k=10)))
print(transform(model.get_nearest_neighbors(word_to_jamo('결제'), k=10)))
```

# 5. Word2Vec

- word2vec으로 구한 벡터는 유사도를 반영한 벡터이기 때문에 연산이 가능하고, Neural net으로 최적의 값을 찾아나감.

- character embedding이 아닌 word embedding

```python
# 간단하게 불용어 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

tokenized_data2 = []
for sentence in tqdm(total_data['reviews'].to_list()):
  tokenized_sentence = mecab.morphs(sentence) # 토큰화
  stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # 불용어 제거
  tokenized_data2.append(stopwords_removed_sentence)
```

```python
print("word2vec용 데이터:", tokenized_data2[0])
print("fasttext용 데이터:", tokenized_data[0])
```

```python
from gensim.models import Word2Vec

# vector_size = 벡터 차원
# window = 컨텍스트 윈도우 크기
# min_count = 단어 최소 빈도 수 제한(빈도가 적은 단어들은 학습하지 않음)
# workers = 학습을 위한 프로세스 수
# sg = 0은 cbow, 1은 skip-gram
model2 = Word2Vec(sentences = tokenized_data2, vector_size = 1000, window = 5, min_count = 5, workers = 4, sg = 0)
```

```python
# 완성된 임베딩 매트릭스의 크기 확인
# 단어 총 개수는 15004개. 벡터 차원은 1000으로 축소됨
model2.wv.vectors.shape
```

# 6. FastText와 Word2Vec 결과 비교

```python
print("FastText 유사도:", transform(model.get_nearest_neighbors(word_to_jamo('남동생'), k=10)))
print("Word2Vec 유사도:", model2.wv.most_similar("남동생"))
```

```python
print("FastText 유사도:", transform(model.get_nearest_neighbors(word_to_jamo('주문'), k=10)))
print("Word2Vec 유사도:", model2.wv.most_similar("주문"))
```

```python
print("FastText 유사도:", transform(model.get_nearest_neighbors(word_to_jamo('배송'), k=10)))
print("Word2Vec 유사도:", model2.wv.most_similar("배송"))
```

```python
print("FastText 유사도:", transform(model.get_nearest_neighbors(word_to_jamo('정치'), k=10)))
print("Word2Vec 유사도:", model2.wv.most_similar("정치"))
```

```python
print("FastText 유사도:", transform(model.get_nearest_neighbors(word_to_jamo('의류'), k=10)))
print("Word2Vec 유사도:", model2.wv.most_similar("의류"))
```

```python
print("FastText 유사도:", transform(model.get_nearest_neighbors(word_to_jamo('좋아'), k=10)))
print("Word2Vec 유사도:", model2.wv.most_similar("좋아"))
```

유사도 점수 측면에서 봤을 때 FastText가 Word2Vec보다 높은 점수대를 띄고 있고 높은 순위에 있는 단어들은 의미와 형태도 근접헤있는 것을 볼 수 있다.

또한 FastText는 단어와 비슷한 형태의 단어들이 높은 유사도 점수를 가지고 있는 반면, Word2Vec의 경우에는 비슷한 의미를 가진 중요 단어들 위주로 높은 유사도 점수를 띄고 있다.

예를 들어 '의류'라는 단어는 Word2Vec에서는 '언더', '아웃도어' 등 관련성이 높은 단어들이 나타나는 반면, FastText의 경우에는 '의뢰', '의의', '의미' 등 '의'라는 어절이 들어간 자주 나타나는 단어들이 유사한 단어들로 출현한다.

또한 분석하는 corpus의 특성에 따라 가장 유사한 단어 모음들을 출력하는 데 있어 단어마다 다른 유의미한 정도를 가진다. 현재 네이버 쇼핑 리뷰를 기본 말뭉치로 하고 있기 때문에 '배송', '주문' 등은 비슷한 단어들이 묶여서 같이 나오지만 쇼핑 리뷰에는 자주 등장하지 않는 '정치'의 경우에는 인간의 상식으로는 이해하기 어려운 단어들이 유사하다고 나오고 있다.
