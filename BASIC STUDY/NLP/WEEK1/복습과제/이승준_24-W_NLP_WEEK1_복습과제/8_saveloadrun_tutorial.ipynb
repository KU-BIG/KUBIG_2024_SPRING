## 모델 저장하고 불러오기

```python
import torch
import torchvision.models as models
```

## 모델 가중치 저장하고 불러오기

학습한 매개변수를 state_dict 라는 내부 상태사전에 저장

```python
model = models.vgg16(weights='IMAGENET1K_V1')
torch.save(model.state_dict(), 'model_weights.pth')
```

불러올때는 동일한 모델의 인스턴스를 생성해 load_state_dict() 메소드를 이용해 불러온다.

```python
model = models.vgg16()
model.load_state_dict(torch.load('model_weights.pth'))
model.eval() #평가 모드로 바꿔야 함.
```

## 모델 형태를 포함하여 저장하고 불러오기

클래스의 구조와 모델까지 함께 저장하기 위해서는
model.state_dict() 가 아닌 model 을 저장함수에 전달

```python
torch.save(model, 'model.pth')
```

```python
#다음과 같이 불러올 수 있다.
model = torch.load('model.pth')
```
